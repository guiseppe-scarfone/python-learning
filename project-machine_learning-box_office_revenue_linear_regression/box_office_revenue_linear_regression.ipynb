# Linear Regression - Box Office Revenue

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
data.head()
```

```python
X = pd.DataFrame(data, columns=['production_budget_usd'])
y = pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwode Gross $')
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope Coefficient:

```python
regression.coef_ # theta_1
```

```python
# Intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.plot(X.values, regression.predict(X.values), color='red', linewidth=5)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwode Gross $')
plt.xlim(0, 450000000)
plt.ylim(0, 3000000000)
plt.show()
```

```python
# R Squared
regression.score(X, y)
```
